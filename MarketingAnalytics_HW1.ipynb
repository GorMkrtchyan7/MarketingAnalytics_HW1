```python
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Nintendo.xlsx')
```

```python
data
```

```python
def bass_model(t, p, q, M):
    return M * (1 - np.exp(-(p+q)*t)) / (1 + (q/p) * np.exp(-(p+q)*t))
```

```python
# Initial parameter guesses
initial_guess = [0.01, 0.1, 10]

# Extract the years and total sold units from your data
years = data['year']
total_units = data['total_sold_units']

# Perform curve fitting
params, covariance = curve_fit(bass_model, years, total_units, p0=initial_guess)

# Extract the estimated parameters
p_estimated, q_estimated, M_estimated = params
```

```python
print("Estimated Coefficient of Innovation (p):", p_estimated)
print("Estimated Coefficient of Imitation (q):", q_estimated)
print("Estimated Market Size (M):", M_estimated)
```

```python
def bass_model(t, p, q, M):
    return M * (1 - np.exp(-(p + q) * t)) / (1 + (q / p) * np.exp(-(p + q) * t))

initial_guess = [0.01, 0.1, 10]

years = data['year']
total_units = data['total_sold_units']

params, covariance = curve_fit(bass_model, years, total_units, p0=initial_guess)

p_estimated, q_estimated, M_estimated = params

future_years = range(min(years), max(years) + 5)
predicted_units = bass_model(future_years, p_estimated, q_estimated, M_estimated)

plt.figure(figsize=(10, 6))
plt.scatter(years, total_units, label='Actual Data', color='blue')
plt.plot(future_years, predicted_units, label='Predicted Data', color='red')
plt.xlabel('Year')
plt.ylabel('Total Sold Units')
plt.title('Bass Model Prediction of Innovation Diffusion')
plt.legend()
plt.grid(True)
plt.show()
```
